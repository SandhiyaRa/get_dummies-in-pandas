## Machine learning models work very well for dataset having only numbers. 
But how do we handle text information in dataset? Simple approach is to use interger or label encoding but when categorical variables are nominal, using simple label encoding can be problematic. One hot encoding is the technique that can help in this situation. In this notebook, we will use pandas get_dummies method to create dummy variables that allows us to perform one hot encoding on given dataset. Alternatively we can use sklearn.preprocessing OneHotEncoder as well to create dummy variables. 

```python
import numpy as np
import pandas as pd

df = pd.read_csv("car-price.csv")
df
```

```python
#Since the Car Model is in categorical type, there are two ways to handle this type of data.
#1. Using get_dummies from pandas
#2. Using One Hot Encoding 

#Thi notebook gives a flovour of both the methods

#Using gtet_dummies: 

dummies =pd.get_dummies(df['Car Model'])
dummies
```

```python
merged = pd.concat([df,dummies],axis='columns')
```

```python
merged
```

```python
final= merged.drop(["Car Model","Mercedez Benz C class"],axis='columns')
final
```

```python
X= final.drop(['Sell Price($)'],axis='columns')
X
```

```python
y= final['Sell Price($)']
y
```

```python
from sklearn import linear_model
regression = linear_model.LinearRegression()
regression.fit(X,y)
```

```python
regression.predict([[45000,4,0,0]])
```

```python

regression.predict([[86000,7,0,1]])
```
